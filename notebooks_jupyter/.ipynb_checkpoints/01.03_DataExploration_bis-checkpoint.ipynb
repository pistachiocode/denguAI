## Import libraries

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler

import dengue_utils as dutils
import dsutils as du
```

```python
def feature_engineering(df):## Add month of the year

    print("1. Including month of the year from 'week_start_date....'")
    
    if 'week_start_date' in df.columns:
        df['week_start_date'] = pd.to_datetime(df['week_start_date'])
        df['monthofyear'] = df['week_start_date'].apply(lambda x: x.month)
        
        df = df.drop(['week_start_date'], axis=1)

    # High correlation between ndvi_nw-ndvi-ne and ndvi-sw-ndvi-se
    # Add the mean of each pair that indicates the level of vegetation in the north and south of both cities.

    # Features engineering
    print("2. Including ndvi_north and  ndvi_south from ndvi_nw, ndvi_ne, ndvi_sw, ndvi_se.... ")
    
    if 'ndvi_nw' in df.columns and 'ndvi_ne' in df.columns and 'ndvi_sw' in df.columns and 'ndvi_se' in df.columns:
    
        df['ndvi_north'] = df[['ndvi_nw', 'ndvi_ne']].mean(axis=1)
        df['ndvi_south'] = df[['ndvi_sw', 'ndvi_se']].mean(axis=1)
        
        df = df.drop(['ndvi_sw', 'ndvi_nw', 'ndvi_ne', 'ndvi_se'], axis=1)

    # Conversion kelvin to celsius
    print("3. Temperature units conversion: kelvin to celsius from some features...")
    #df = dutils.temperature_conversion(df)
    
    # City to numeric
    
  #  df['city_code'] = np.where(df['city'] == 'sj', 0, 1)
    # Drop temperature features
    
   # df = df.drop(['precipitation_amt_mm', 'station_avg_temp_c', 'station_max_temp_c', 
   #               'station_min_temp_c', 'station_precip_mm', 'reanalysis_tdtr_k'], axis=1)
   # 
    return df
```

```python
# Load dataset

features_train = pd.read_csv('dengue_features_train.csv')
labels_train   = pd.read_csv('dengue_labels_train.csv')

features_test = pd.read_csv('dengue_features_test.csv')

# Join train labes and features
input_train_dataset = features_train.merge(labels_train, 
                                           left_on = ['city', 'year', 'weekofyear'], 
                                           right_on = ['city', 'year', 'weekofyear'],
                                           how = 'inner')

```

```python
input_train_dataset.head()
```

```python
input_train_dataset = feature_engineering(input_train_dataset)
```

```python
input_train_dataset = input_train_dataset.fillna(method='ffill').drop_duplicates()

cities = input_train_dataset['city'].unique()

for city in cities:
    
    train_dataset_city = input_train_dataset[input_train_dataset['city'].isin([city])]    
        
    du.show_feature_correlation(train_dataset_city, 'total_cases', title= "Correlation features city: " + city)
```

Exploration for each city

```python
for city in cities:
    
    # 1. Filter city
    train_dataset_city = input_train_dataset[input_train_dataset['city'].isin([city])]    
    du.printlog("Preprocessing " + city + " size: "  + str(train_dataset_city.shape))
    
    # 2. Drop city column
    train_dataset_city = train_dataset_city.drop(['city'], axis=1)

    # 3. Scale data
    dutils.printlog("Scale dataset: " + str(train_dataset_city.shape)) 
    
    scaler = StandardScaler().fit(train_dataset_city)
    train_dataset_city = pd.DataFrame(scaler.transform(train_dataset_city), columns = train_dataset_city.columns)
    
    # 4. Including old weather data

    climate_variables = train_dataset_city.columns.drop(['weekofyear', 'total_cases', 'monthofyear', 'year'])

    for j, climate_variable in enumerate(climate_variables):
        train_dataset_city[climate_variable] = (train_dataset_city
                                                .sort_values(['year', 'weekofyear'], ascending=True)
                                                .groupby(['year'])[climate_variable]
                                                .shift(8))
    train_dataset_city = train_dataset_city.dropna()
    dutils.printlog("Shift dataset: " + str(train_dataset_city.shape)) 
        
    du.show_feature_correlation(train_dataset_city, 'total_cases', title= "Correlation features city: " + city)
```

```python
import matplotlib.pyplot as plt

corr_df = pd.DataFrame()
    
input_train_dataset = input_train_dataset[input_train_dataset['city'] == 'sj']

climate_variables = input_train_dataset.columns.drop(['weekofyear', 'total_cases', 'city', 'monthofyear', 'year'])
nlags = 12

for j, climate_variable in enumerate(climate_variables):
    print(climate_variable)
    df_test_lag = input_train_dataset[['city', 'year', 'weekofyear', climate_variable, 'total_cases']]

    for i in np.arange(1, nlags-1, 1):
        df_test_lag[climate_variable + str(i)] = (df_test_lag
                                                        .sort_values(['city', 'year', 'weekofyear'], ascending=True)
                                                        .groupby(['city', 'year'])[climate_variable]
                                                        .shift(i))
        df_test_lag = df_test_lag.drop_duplicates().dropna()

        corr_df_lag = pd.DataFrame(df_test_lag.corr()['total_cases'])
        corr_df_lag['nrows'] = df_test_lag.shape[0]
        corr_df_lag['lag'] = i
        #print(corr_df_lag.columns)
        corr_df = pd.concat([corr_df, corr_df_lag], axis=0)

    #ax = fig.add_subplot(20, 1, j+1)
    #du.show_feature_correlation(df_test_lag, 'total_cases')
```
